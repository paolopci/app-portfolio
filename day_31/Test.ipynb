# Create a price histogram for wines that cost less than 100

```python
import matplotlib.pyplot as plt
```

```python
# 3. Filtra i vini con prezzo inferiore a 100
df_filtered = df[df['price'] < 100]
```

```python
# 4. Crea l'istogramma dei prezzi
plt.figure(figsize=(10, 6))
plt.hist(df_filtered['price'], bins=20, edgecolor='black')  # puoi modificare il numero di bin
plt.title('Istogramma dei prezzi dei vini (< 100)')
plt.xlabel('Prezzo')
plt.ylabel('Numero di vini')
plt.grid(True)
plt.tight_layout()
plt.show()
```
